```python
import pandas as pd
```

Download datasets to get the users with the songs they listen to and info about each song such as gender of artist and genre

```python
# Download unique_tracks dataset that includes song data (track_id, song_id, artist, and song)
data_song = pd.read_csv("data/unique_tracks.txt", engine='python', sep="<SEP>", header = None, names=["track_id", "song_id", "artist", "song"])
data_song
```

```python
# Download train_triplets dataset with 100,000 songs that includes user data (user_id, song_id, play_count)  
data_user = pd.read_csv("data/train_triplets.txt", sep="\t", header = None, names=["user_id", "song_id", "play_count" ], nrows = 100000)
data_user
```

```python
# Merge the two datasets and drop the songs if they don't have user data to go with it
song_df = pd.merge(data_user, data_song.drop_duplicates(['song_id']), on="song_id", how="inner")
song_df
```

```python
# Download gender_dataset dataset that includes data about the artist (chartmetric_id, name, artist_country, pronoun, gender, is_band, genre)
gender_data = pd.read_csv("data/Gender_Dataset_1.7.csv")
gender_data.head()
```

```python
# This could be used to remove all bands from data
#gender_data.drop(gender_data[gender_data['is_band'] == True].index, inplace = True)
```

```python
# Merge the datasets so that the gender data is included
full_df = song_df.merge(gender_data,left_on='artist', right_on='name', how="inner") 
full_df
```

Once the datasets are merged, it is time to clean the dataset and prepare it for running the recommender systems and the analysis

```python
# Edit the genre column so that each genre is seperated by ','
genre_data = full_df['genre'].str.split(',', expand = True)
full_df['genre'] = full_df['genre'].str.replace('{', '')
full_df['genre'] = full_df['genre'].str.replace('}', '')
full_df['genre'] = full_df['genre'].str.replace('"', '')
full_df
```

```python
# the genre column so that there are separate columns for each genre
full_df['genre'].str.split(',', expand = True)

# Seperate the first three genres into new columns
full_df['genre_1'] = genre_data[0]
full_df['genre_2'] = genre_data[1]
full_df['genre_3'] = genre_data[2]
full_df
```

```python
full_df['genre_1'] = full_df['genre_1'].str.replace('{', '')
full_df['genre_1'] = full_df['genre_1'].str.replace('}', '')
full_df['genre_1'] = full_df['genre_1'].str.replace('"', '')
full_df['genre_2'] = full_df['genre_2'].str.replace('{', '')
full_df['genre_2'] = full_df['genre_2'].str.replace('}', '')
full_df['genre_2'] = full_df['genre_2'].str.replace('"', '')
full_df['genre_3'] = full_df['genre_3'].str.replace('{', '')
full_df['genre_3'] = full_df['genre_3'].str.replace('}', '')
full_df['genre_3'] = full_df['genre_3'].str.replace('"', '')
full_df
```

```python
# Drop the columns that are not needed for the analysis
full_df = full_df.drop(columns = ["genre", "chartmetric_id", "is_band", "name", "track_id", "gender", "artist_country"])
```

```python
# Save this dataset to be used in the analysis portion
full_df.to_csv('full_dataset.csv')
full_df.head()
```

```python
# Create a column in the dataset to include the song, artist, pronouns, and genres
# This will be used so that all the song information is present after running the recommender systems
full_df['song'] = full_df['song'].map(str) + " ~ " + full_df['artist'] + " ~ " + full_df['pronoun'] + " ~ " + full_df['genre_1'] + " ~ " + full_df['genre_2'] + " ~ " + full_df['genre_3']
full_df = full_df.drop(columns = ["artist"])
full_df = full_df.dropna()
full_df.head()
```

```python
# Save this dataset to use for the recommender systems
full_df.to_csv('full_df.csv')
```
